```python
#  Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

#  Load the Iris dataset from seaborn (built-in)
iris = sns.load_dataset('iris')

#  Show the first few rows of the dataset
print("Head of the dataset:")
print(iris.head())

#  Check for null values
print("\nNull values in the dataset:")
print(iris.isnull().sum())

#  Statistical Calculations
mean_sepal_length = iris['sepal_length'].mean()
median_sepal_width = iris['sepal_width'].median()
mode_petal_length = iris['petal_length'].mode()[0]  # mode() returns a Series

print(f"\nMean Sepal Length: {mean_sepal_length}")
print(f"Median Sepal Width: {median_sepal_width}")
print(f"Mode Petal Length: {mode_petal_length}")

#  Correlation Heatmap (e.g., sepal_length and petal_length)
plt.figure(figsize=(6, 4))
sns.heatmap(iris[['sepal_length', 'petal_length']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation: Sepal Length vs Petal Length")
plt.show()

#  Additional Visualizations

# 1. Pairplot to visualize all pairwise relationships
sns.pairplot(iris, hue='species')
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()

# 2. Boxplot for Sepal Width across species
plt.figure(figsize=(6, 4))
sns.boxplot(x='species', y='sepal_width', data=iris, palette='Set2')
plt.title("Boxplot of Sepal Width by Species")
plt.show()

# 3. Violin plot for Petal Length
plt.figure(figsize=(6, 4))
sns.violinplot(x='species', y='petal_length', data=iris, palette='Set3')
plt.title("Violin Plot of Petal Length by Species")
plt.show()

# 4. Histogram of Sepal Length
plt.figure(figsize=(6, 4))
plt.hist(iris['sepal_length'], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```
